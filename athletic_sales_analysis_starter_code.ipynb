```python
# Import Libraries and Dependencies
import pandas as pd
import numpy as np
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
raw_df1 = pd.read_csv("../Resources/athletic_sales_2020.csv")
raw_df2 = pd.read_csv("../Resources/athletic_sales_2021.csv")

```

```python
# Display the 2020 sales DataFrame
raw_df1
```

```python
# Display the 2021 sales DataFrame
raw_df2
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
raw_df1.dtypes
```

```python
# Check the 2021 sales data types.
raw_df2.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_df = pd.concat([raw_df1, raw_df2], axis = 0, join='inner').reset_index(drop=True)
combined_df
```

```python
# Check if any values are null.
if combined_df.isnull().values.any():
    print("There are null values in the DataFrame.")
else:
    print("There are no null values in the DataFrame.")
```

```python
# Check the data type of each column
combined_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
# combined_df['invoice_date'] = pd.to_datetime(combined_df['invoice_date'], format='%d-%m-%Y')
combined_df['invoice_date'] = pd.to_datetime(combined_df['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_df
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
#Total_Products_Sold = combined_df.groupby(['region', 'state', 'city']).agg({"units_sold":"sum"})
Total_Products_Sold = combined_df.groupby(['region', 'state', 'city'])["units_sold"].sum()
#Total_Products_Sold
Total_Products_Sold_sorted = Total_Products_Sold.sort_values(ascending=False)
# Show the top 5 results.
Total_Products_Sold_sorted.head()                           
```

```python
# Show the number products sold for region, state, and city.
Products_sold = pd.pivot_table(combined_df, values='units_sold', index=['retailer'], columns=['region', 'state', 'city'], aggfunc=np.sum)
Products_sold
# Rename the "units_sold" column to "Total_Products_Sold"
renamed_column = combined_df.rename(columns={'units_sold': 'Total_Products_Sold'})

# Show the top 5 results.
renamed_column_sorted = renamed_column.sort_values(by='Total_Products_Sold', ascending=False)
renamed_column_sorted.head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
# women_athletic_footwear_sales_data = pd.pivot_table(combined_df, values='total_sales', index=['retailer'], columns=['region', 'state', 'city'], aggfunc=np.sum)
# df.loc[(df["column_name"] =="<value>")]
women_athletic_footwear_sales_data = combined_df.loc[(combined_df["product"] == "Women's Athletic Footwear")]
women_athletic_footwear_sales_data.head()
```

#### Using `groupby`

```python
## Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
Total_number_of_women_athletic_sales = women_athletic_footwear_sales_data.groupby(['retailer', 'region', 'state', 'city']).agg({"units_sold":"sum"}) 
Total_number_of_women_athletic_sales
#women_atheletic_footwear_sales = combined_df[combined_df["product"] == "Women's Athletic Footwear"]

## Group by "retailer", "region", "state", and "city" columns and sum the total sales
grouped_sales = women_athletic_footwear_sales_data.groupby(["retailer", "region", "state", "city"]).agg({"total_sales":"sum"})
grouped_sales
## women_atheletic_footwear_sales_data_per_rsc  = pd.pivot_table(combined_df, values='total_sales', index=['retailer'], columns=['region', 'state', 'city'], aggfunc=np.sum)
## Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
renamed_units_sold = Total_number_of_women_athletic_sales.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})
renamed_units_sold
## Show the top 5 results.
renamed_units_sold_sorted = renamed_units_sold.sort_values(by="Womens_Footwear_Units_Sold", ascending = False)
renamed_units_sold_sorted.head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Group by "retailer", "region", "state", and "city" columns and aggregate the total sales
# women_footwear_sales_summary = women_atheletic_footwear_sales_data.groupby(["region", "state", "city"]).size().reset_index(name="total_sales")
total_number_of_womens_athletic_footwear_sold = women_athletic_footwear_sales_data.pivot_table(index=["retailer", "region", "state", "city"], values="total_sales", aggfunc="sum")
total_number_of_womens_athletic_footwear_sold
# Sort the results in descending order based on total sales
total_number_of_womens_athletic_footwear_sold_sorted = total_number_of_womens_athletic_footwear_sold.sort_values(by="total_sales", ascending=False)
# women_footwear_sales_summary = women_footwear_sales
# Display the DataFrame with total sales for each combination of retailer, region, state, and city
total_number_of_womens_athletic_footwear_sold_sorted

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
renamed_units_sold = Total_number_of_women_athletic_sales.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})
renamed_units_sold
# Show the top 5 results.
renamed_units_sold_sorted = renamed_units_sold.sort_values(by="Womens_Footwear_Units_Sold", ascending = False)
renamed_units_sold_sorted.head(5)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
sales_table = pd.pivot_table(women_athletic_footwear_sales_data, index='invoice_date', values='total_sales', aggfunc="sum")
sales_table

# Optional: Rename the "total_sales" column to "Total Sales"
renamed_total_sales = sales_table.rename(columns={'total_sales': 'Total Sales'})
# Show the table.
renamed_total_sales_sorted = renamed_total_sales.sort_values(by="Total Sales", ascending=False)
renamed_total_sales_sorted.head(10)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_sales = sales_table.resample('D').sum()
daily_sales
renamed_total_sales = daily_sales.rename(columns={'total_sales': 'Total Sales'})
# Sort the resampled pivot table in descending order on "Total Sales".
daily_sales_sorted = renamed_total_sales.sort_values(by='Total Sales', ascending=False)
daily_sales_sorted.head(10)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales = renamed_total_sales.resample('W').sum()
weekly_sales
# Sort the resampled pivot table in descending order on "Total Sales".
weekly_sales_sorted = weekly_sales.sort_values(by='Total Sales', ascending=False)
weekly_sales_sorted.head(10)
```
